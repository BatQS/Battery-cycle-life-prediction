```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import ElasticNetCV, ElasticNet
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
df = pd.read_csv('12features.csv')
df

# --- 1. Data Preparation ---
# Assume 'df' is your pandas DataFrame
y = df['observe'].values          # Target values (what we want to predict)
X = df.iloc[:, :-1].values        # Feature matrix (the first 17 columns)

# Feature Scaling: Bringing all features to the same scale (Crucial for ElasticNet!)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 2. Automatic Hyperparameter Tuning (CV) ---
# ElasticNetCV automatically tests different alpha and l1_ratio values 
# to find the best combination using Cross-Validation.
cv_model = ElasticNetCV(
    l1_ratio=[.1, .5, .7, .9, .95, 1], 
    cv=5, 
    max_iter=100000
)
cv_model.fit(X_scaled, y)

print(f"Optimal Penalty (alpha): {cv_model.alpha_:.4f}")
print(f"Optimal Mix Ratio (l1_ratio): {cv_model.l1_ratio_:.4f}")

# --- 3. Build Final Model with Best Parameters ---
final_model = ElasticNet(
    alpha=cv_model.alpha_, 
    l1_ratio=cv_model.l1_ratio_
)
final_model.fit(X_scaled, y)

# --- 4. Model Evaluation (Cross-Validation Prediction) ---
# We use cross_val_predict to get "out-of-sample" predictions for a fair evaluation.
y_pred = cross_val_predict(final_model, X_scaled, y, cv=4)

# --- 5. Calculate Performance Metrics ---
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print(f"Evaluation Results: MAE={mae:.2f}, RMSE={rmse:.2f}, R2={r2:.2f}")

# --- 6. Visualization ---
plt.figure(figsize=(6, 6))
# Plotting the predicted values vs actual values
plt.scatter(y, y_pred, color='#4cb391', alpha=0.7, label='Predictions') 

# Drawing a 45-degree reference line (Perfect Prediction Line)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2, label='Ideal Fit') 

plt.xlabel("Actual Value (Experimental)")
plt.ylabel("Predicted Value")
plt.title(f"ElasticNet Regression Results (R²={r2:.2f})")
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

# generate the CR_10:100	intercept_DQ91:100	slope_DQ2:100 and find the cycle when the battery capacity drop to 80% of the initial cycle capacity

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ===== 1. Read CSV file =====
file_path = "Li金属電池_サイクル特性.CSV"
df = pd.read_csv(file_path, header=None, encoding='latin1')

# ===== 2. Extract battery names and actual data =====
battery_names = df.iloc[0, :].tolist()
data = df.iloc[2:, :].reset_index(drop=True)

# ===== 3. Define helper functions =====
def get_capacity_at_cycle(cycle_series, capacity_series, target_cycle):
    """
    Extract the capacity value corresponding to a specific cycle number.

    Parameters
    ----------
    cycle_series : pandas.Series
        Series containing cycle numbers.
    capacity_series : pandas.Series
        Series containing capacity values.
    target_cycle : int
        The target cycle number to search for.

    Returns
    -------
    float
        Capacity value at the specified cycle.
        Returns NaN if the cycle does not exist.
    """
    mask = cycle_series == target_cycle
    values = capacity_series[mask]
    if len(values) == 0:
        return np.nan
    return values.iloc[0]


def find_cycle_at_80_percent_discrete(cycle, capacity):
    """
    Find the first discrete cycle at which the battery capacity
    drops to or below 80% of the initial capacity.

    The initial capacity is defined as the capacity at cycle 1.

    Parameters
    ----------
    cycle : pandas.Series
        Series containing cycle numbers.
    capacity : pandas.Series
        Series containing capacity values.

    Returns
    -------
    int or float
        The first cycle number where capacity <= 80% of initial capacity.
        Returns NaN if the battery never reaches this threshold.
    """
    Q_initial = get_capacity_at_cycle(cycle, capacity, 1)
    if pd.isna(Q_initial):
        return np.nan

    threshold = 0.8 * Q_initial
    below = capacity <= threshold

    if not below.any():
        return np.nan

    idx = np.argmax(below)
    return cycle.iloc[idx]


def find_cycle_at_80_percent_interp(cycle, capacity):
    """
    Estimate the cycle number at which the battery capacity
    reaches 80% of the initial capacity using linear interpolation.

    The initial capacity is defined as the capacity at cycle 1.
    This method provides a more accurate estimate than the discrete method
    by interpolating between two neighboring data points around the threshold.

    Parameters
    ----------
    cycle : pandas.Series
        Series containing cycle numbers.
    capacity : pandas.Series
        Series containing capacity values.

    Returns
    -------
    float
        Interpolated cycle number corresponding to 80% of initial capacity.
        Returns NaN if the battery never reaches this threshold.
    """
    Q_initial = get_capacity_at_cycle(cycle, capacity, 1)
    if pd.isna(Q_initial):
        return np.nan

    threshold = 0.8 * Q_initial

    if capacity.iloc[0] <= threshold:
        return cycle.iloc[0]

    for i in range(1, len(capacity)):
        y1 = capacity.iloc[i - 1]
        y2 = capacity.iloc[i]
        x1 = cycle.iloc[i - 1]
        x2 = cycle.iloc[i]

        if (y1 > threshold and y2 <= threshold) or (y1 < threshold and y2 >= threshold):
            if y2 == y1:
                return x2

            cycle_interp = x1 + (threshold - y1) * (x2 - x1) / (y2 - y1)
            return cycle_interp

    return np.nan


# ===== 4. Prepare plotting =====
plt.figure(figsize=(10, 6))

num_batteries = data.shape[1] // 2
colors = plt.cm.viridis(np.linspace(0, 1, num_batteries))

# ===== 5. Extract features for each battery =====
results = []

for i in range(0, data.shape[1], 2):
    # Each battery consists of two columns:
    # first column = cycle number, second column = discharge capacity
    cycle = pd.to_numeric(data.iloc[:, i], errors='coerce')
    capacity = pd.to_numeric(data.iloc[:, i + 1], errors='coerce')

    # Remove NaN values
    mask = cycle.notna() & capacity.notna()
    cycle = cycle[mask].reset_index(drop=True)
    capacity = capacity[mask].reset_index(drop=True)

    # Convert cycle values to integers if possible
    try:
        cycle = cycle.astype(int)
    except:
        pass

    label = str(battery_names[i])

    # ===== Plot capacity vs cycle number =====
    plt.plot(
        cycle,
        capacity,
        linewidth=2,
        color=colors[i // 2],
        label=label
    )

    # Extract capacity values at selected cycles
    Q1 = get_capacity_at_cycle(cycle, capacity, 1)
    Q2 = get_capacity_at_cycle(cycle, capacity, 2)
    Q10 = get_capacity_at_cycle(cycle, capacity, 10)
    Q91 = get_capacity_at_cycle(cycle, capacity, 91)
    Q100 = get_capacity_at_cycle(cycle, capacity, 100)

    # ===== Feature 1: CR_10:100 =====
    if pd.notna(Q10) and Q10 != 0 and pd.notna(Q100):
        CR_10_100 = Q100 / Q10
    else:
        CR_10_100 = np.nan

    # ===== Feature 2: intercept_DQ91:100 =====
    if pd.notna(Q91) and pd.notna(Q100):
        slope_91_100 = (Q100 - Q91) / (100 - 91)
        intercept_DQ91_100 = Q91 - slope_91_100 * 91
    else:
        slope_91_100 = np.nan
        intercept_DQ91_100 = np.nan

    # ===== Feature 3: slope_DQ2:100 =====
    if pd.notna(Q2) and pd.notna(Q100):
        slope_DQ2_100 = (Q100 - Q2) / (100 - 2)
    else:
        slope_DQ2_100 = np.nan

    # ===== Cycle number at 80% of initial capacity =====
    cycle_80_discrete = find_cycle_at_80_percent_discrete(cycle, capacity)
    cycle_80_interp = find_cycle_at_80_percent_interp(cycle, capacity)

    results.append({
        "Battery": label,
        "Q1": Q1,
        "Q2": Q2,
        "Q10": Q10,
        "Q91": Q91,
        "Q100": Q100,
        "CR_10:100": CR_10_100,
        "intercept_DQ91:100": intercept_DQ91_100,
        "slope_DQ2:100": slope_DQ2_100,
        "Cycle_at_80%_discrete": cycle_80_discrete,
        "Cycle_at_80%_interp": cycle_80_interp
    })

# ===== 6. Finalize plot =====
plt.xlabel("Cycle Number", fontsize=14)
plt.ylabel("Discharge Capacity (mAh)", fontsize=14)
plt.title("Capacity vs. Cycle Number", fontsize=16)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=9)
plt.tight_layout()
plt.savefig("capacity_cycle_num.jpeg", bbox_inches='tight')
plt.show()

# ===== 7. Convert results to DataFrame =====
results_df = pd.DataFrame(results)

# ===== 8. Print results =====
print(results_df)

# ===== 9. Save results =====
results_df.to_csv("battery_features_and_80pct_cycle.csv", index=False, encoding="utf-8-sig")
```

# generate the 	logminDQ100-10	logvarDQ100-10 features

```python
import pandas as pd
import numpy as np
import os

# ==========================================
# Folder containing your CSV files
# ==========================================
folder_path = "/Users/siqianli/Desktop/260316_都立大_Li金属電池データ"   

# Automatically find all CSV files
file_list = [f for f in os.listdir(folder_path) if f.lower().endswith('.csv')]

print(f"Found {len(file_list)} CSV files")

# ==========================================
# Helper function for interpolation
# ==========================================
def prepare_for_interp(capacity, voltage):
    sort_idx = np.argsort(voltage)
    voltage_sorted = voltage[sort_idx]
    capacity_sorted = capacity[sort_idx]

    voltage_unique, unique_idx = np.unique(voltage_sorted, return_index=True)
    capacity_unique = capacity_sorted[unique_idx]

    return capacity_unique, voltage_unique

# ==========================================
# Store results
# ==========================================
results = []

# ==========================================
# Loop over all files
# ==========================================
for file_name in file_list:

    file_path = os.path.join(folder_path, file_name)

    try:
        df = pd.read_csv(file_path, encoding='latin1')

        curves = []

        # Extract curves
        for i in range(2, df.shape[1] - 1, 4):
            capacity = pd.to_numeric(df.iloc[:, i], errors='coerce')
            voltage = pd.to_numeric(df.iloc[:, i + 1], errors='coerce')

            mask = capacity.notna() & voltage.notna()
            capacity = capacity[mask].values
            voltage = voltage[mask].values

            if len(capacity) < 2:
                continue

            curves.append((capacity, voltage))

        # Check if enough cycles
        if len(curves) < 100:
            results.append({
                "file_name": file_name,
                "num_valid_cycles": len(curves),
                "log_min_qd_diff": np.nan,
                "log_var_qd_diff": np.nan,
                "status": "Not enough cycles"
            })
            continue

        # Get cycle 10 and 100
        cap_10, volt_10 = curves[9]
        cap_100, volt_100 = curves[99]

        # Common voltage range
        v_min = max(np.min(volt_10), np.min(volt_100))
        v_max = min(np.max(volt_10), np.max(volt_100))

        if v_min >= v_max:
            results.append({
                "file_name": file_name,
                "num_valid_cycles": len(curves),
                "log_min_qd_diff": np.nan,
                "log_var_qd_diff": np.nan,
                "status": "No overlap"
            })
            continue

        voltage_grid = np.linspace(v_min, v_max, 1000)

        # Prepare interpolation
        cap10_u, volt10_u = prepare_for_interp(cap_10, volt_10)
        cap100_u, volt100_u = prepare_for_interp(cap_100, volt_100)

        # Interpolate
        Q10_interp = np.interp(voltage_grid, volt10_u, cap10_u)
        Q100_interp = np.interp(voltage_grid, volt100_u, cap100_u)

        # ΔQ
        Qd_diff = Q100_interp - Q10_interp

        # Features
        eps = 1e-12
        log_min = np.log10(np.abs(np.nanmin(Qd_diff)) + eps)
        log_var = np.log10(np.abs(np.nanvar(Qd_diff)) + eps)

        results.append({
            "file_name": file_name,
            "num_valid_cycles": len(curves),
            "log_min_qd_diff": log_min,
            "log_var_qd_diff": log_var,
            "status": "Success"
        })

    except Exception as e:
        results.append({
            "file_name": file_name,
            "num_valid_cycles": np.nan,
            "log_min_qd_diff": np.nan,
            "log_var_qd_diff": np.nan,
            "status": f"Error: {str(e)}"
        })

# ==========================================
# Save results
# ==========================================
results_df = pd.DataFrame(results)

output_path = os.path.join(folder_path, "summary_qd_diff_features.csv")
results_df.to_csv(output_path, index=False, encoding="utf-8-sig")

print(results_df)
print(f"\nSaved to: {output_path}")
```
